```python
import pandas as pd

# Load the dataset
df = pd.read_excel('Road Accident fatility dataset.xlsx')

# Display first five rows
df.head()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Set style for plots
plt.style.use('ggplot')
sns.set_theme(style = 'whitegrid')
```

```python
# Dataset information
print("data Info:")
df.info()

# Descriptive statistics
print("\nDescriptive Statistics:")
df.describe()

# Checking missing values
print("\nMissing Values:")
df.isnull().sum()
```

```python
# Statistical summary of the dataset
df.describe()
```

```python
# Import required libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Year-wise road accident fatalities distribution
years = ['2020', '2021', '2022', '2023', '2024']
totals = df[years].sum()

# Create visualization
plt.figure(figsize=(8,5))
sns.barplot(x=years, y=totals.values)

# Add labels and title
plt.title("Year-wise Road Accident Fatalities")
plt.xlabel("Year")
plt.ylabel("Number of Fatalities")

# Display graph
plt.show()
```

```python
# Import required libraries
import matplotlib.pyplot as plt

# remove total row
data = df[df['State/UT'] != 'Total (India)']

# Plot bar chart
plt.figure(figsize=(14,6))
plt.bar(data['State/UT'], data['2024'])

# Labels and title
plt.title("State-wise Road Accident Fatalities (2024)")
plt.xlabel("States/UTs")
plt.ylabel("Fatalities")

# Rotate labels
plt.xticks(rotation=90)

# Adjust layout
plt.tight_layout()

# Show graph
plt.show()
```

```python
# Import required libraries
import matplotlib.pyplot as plt

# Remove total row if present 
data = df[df['State/UT'] != 'Total (India)']

# Select to 10 states with highest fatalitiesin 2024
top_states = data.sort_values(by='2024', ascending=False).head(10)

# Create bar chart
plt.figure(figsize=(10,6))
plt.bar(top_states['State/UT'], top_states['2024'])

# Add title and labels
plt.title("Top 10 States with Highest Road Acident fatalities (2024)")
plt.xlabel("States/UTs")
plt.ylabel("Number of Fatalities")

# Adjust x-axis labels
plt.xticks(rotation=45)

# Adjust layout
plt.tight_layout()

# Display graph
plt.show()
```

```python
import matplotlib.pyplot as plt

# Top 10 lowest states
low_states = df.nsmallest(10, '2024')

# Figure size
plt.figure(figsize=(10, 6))

# Bar chart
bars = plt.bar(
    low_states['State/UT'],
    low_states['2024'],
    color='steelblue',
    edgecolor='navy'
)

plt.title("Top 10 Lowest Road Accident Fatalities (2024)")
plt.xlabel("States/UTs")
plt.ylabel("Fatalities")
plt.xticks(rotation=45, ha='right')

# Values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 2,
        f'{int(height)}',
        ha='center'
    )

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Total fatalities year-wise
totals = df[['2020', '2021', '2022', '2023', '2024']].sum()

# Line chart
plt.plot(totals.index, totals.values, marker='o')
plt.title("Year-wise Road Accident Fatalities Trend")
plt.xlabel("Year")
plt.ylabel("Total fatalities")
plt.show()
```

```python
# Import required libraries
import seaborn as sns 
import matplotlib.pyplot as plt

# Select yearly columns
heatmap_data = df[['2020', '2021', '2022', '2023', '2024']]

# Create heatmap
plt.figure(figsize=(8,5))
sns.heatmap(heatmap_data.corr(), annot=True)

# Add title
plt.title("Correlation Heatmap of Accident fatalities")

# Show graph
plt.show()
```

```python
# Import required libraries
import matplotlib.pyplot as plt

# Compare two states
state1 = df[df['State/UT'] == 'Uttar Pradesh']
state2 = df[df['State/UT'] == 'Tamil Nadu']

years = ['2020', '2021', '2022', '2023', '2024']

# Plot comparison
plt.figure(figsize=(8,5))
plt.plot(years, state1[years].values.flatten(), marker='o', label='Uttar Pradesh')
plt.plot(years, state2[years].values.flatten(), marker='o', label='Tamil Nadu')

# Labels and title
plt.title("Fatalities Comparison")
plt.xlabel("Year")
plt.ylabel("Fatalities")

# Legend
plt.legend()

# Show graph
plt.show()
```
